#Causal impact test on Colab

```r
install.packages("CausalImpact")
library(CausalImpact)
```

## Simulation data preparation

x: 100 samples of simple AR process data generation

```r
set.seed(1)
x1 <- 100 + arima.sim(model = list(ar = 0.999), n = 100)
```

y: simulated　data lifted by promotion by 10

```r
y <- 1.2 * x1 + rnorm(100)
y[71:100] <- y[71:100] + 10
data <- cbind(y, x1)
```

```r
head(data)
```

```r
matplot(data, type = "l")
```

```r
pre.period <- c(1, 70)
post.period <- c(71, 100)
```

## Causal impact

```r
impact <- CausalImpact(data, pre.period, post.period)
```

```r
plot(impact)
```
